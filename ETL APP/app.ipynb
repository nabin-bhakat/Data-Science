```python
import pandas as pd
```

# Extraction

```python
df=pd.read_csv('./data/source/cust_info.csv')
df
```

# Transformation

```python
# dropping all the duplicate values
df=df.drop_duplicates()
```

```python
#removing '.0' from cst_id 
df['cst_id']=df['cst_id'].astype(str)
df['cst_id']=df['cst_id'].str.replace('.0','')
df
```

```python
#removing any special character and spaces from cst_firstname and cst_lastname
df['cst_firstname']=df['cst_firstname'].str.replace('[^a-zA-z]','',regex=True)
df['cst_lastname']=df['cst_lastname'].str.replace('[^a-zA-z]','',regex=True)
```

```python
#standarizred the cst_martial_status column
df['cst_marital_status']=df['cst_marital_status'].str.replace('M','Married')
df['cst_marital_status']=df['cst_marital_status'].str.replace('S','Single')

df['cst_marital_status']
```

```python
#standarizred the cst_gndr column
df['cst_gndr']=df['cst_gndr'].str.replace('M','Male')
df['cst_gndr']=df['cst_gndr'].str.replace('F','Female')

df['cst_gndr']
```

# Loading

```python
df.to_csv('./data/destination/clean.csv',index=False)
```
